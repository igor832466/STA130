### (1)

``df.dropna()`` might be preferred over using ``del df['col']`` when, for example, when one wants to disregard observations that, because of missing data, may skew or disrupt data analysis. 

Using titanic.csv, an example of a "use case" in which using ``df.dropna()`` may be preffered is if one wanted to calculate the average age of all passengers. There are 177 missing age entries which interfere with finding the average age of all 890 passengers. This is because calculating the average age would not calculate the average age of 890 passengers, but rather the average age of 713 of those passengers. Using ``del df['col']`` would delete the entire age column as it has missing entries. However, the goal is to calculate the average age, and as such removing rows with missing entries will produce a more accurate picture.

``df.groupby("col1")["col2"].describe()``

``df.groupby("col1")`` groups DataFrame (df) by the values of a specified column (``"col1"``)

``["col2"]`` selects another specified column (``"col2"``) from each group after grouping by ``"col1"``

``.describe()`` provides summary statistics defined in question 6.

In other words, ``df.groupby("col1")["col2"].describe()`` groups data by ``"col1"`` and, for each kind of statistics in that column, provide summary statistics based on ``"col2"``

```python
# Example: Group columns "fare" by "class"

df.groupby("class")["age"].describe()
```

Over here, passengers have been grouped by their class and have been, for each class, provided summary statistics based on their ``"age"`` column.
